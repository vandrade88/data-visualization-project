# LOAD
Load DataFrames into database

```python
rds_connection_string = "postgres:postgres@localhost:5432/whr_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
df2020.to_sql(name='whr2020', con=engine, if_exists='append', index=True)
```

```python
df2019.to_sql(name='whr2019', con=engine, if_exists='append', index=True)
```

```python
df2018.to_sql(name='whr2018', con=engine, if_exists='append', index=True)
```
